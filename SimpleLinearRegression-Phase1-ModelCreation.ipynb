```python
#Importing pandas library and giving it an alias as pd.Pandas library is mainly used for database table structured data.
import pandas as pd
```

```python
#We are now assigning all the datas of Salary_Data.csv file to a variable called dataset using read_csv function which is under padas library.
dataset=pd.read_csv("Salary_Data.csv")
```

```python
#Checking the value of dataset variable.
dataset
```

```python
#We are saving YearsExperience column datas into independent variable & Salary column datas into dependent variable
independent=dataset[["YearsExperience"]]
dependent=dataset[["Salary"]] 
```

```python
#Checking the value of independent variable.
independent
```

```python
#Checking the value of dependent variable.
dependent
```

```python
#Importing train_test_split function from sklearn library.Importing only the required function without importing all the sk learn library.
from sklearn.model_selection import train_test_split
#Calling train_test_split function with independent and dependent variable and Output gives you 4 array values.2 for training set and 2 for test set.
#Assiging the 4 array output as X_train,X_test,Y_train,Y_test
X_train,X_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)

```

```python
#Importing LinearRegression function from sklearn library
from sklearn.linear_model import LinearRegression
#Assiging the LinearRegression function to a variable called regressor.Since the LinearRegression function is assigned to variable called regressor 
#LinearRegression function behaves as a class.
regressor=LinearRegression()
#We are now calling fit function through regressor variable passing the training dataset so that the model gets created.
#The model is gets created using the fit function means the W=weight & b=bias gets calculated.
regressor.fit(X_train,y_train)
```

```python
#regressor.coef_ is the place where calculated weight is stored for this dataset which is assiged to a variable weight
weight=regressor.coef_
```

```python
#printing the weight variable value
weight
```

```python
#regressor.intercept_ is the place where calculated bias is stored for this dataset which is assiged to a variable bias
bias=regressor.intercept_
```

```python
#printing the bias variable value
bias
```

```python
#We are now predicting the output using the input test data.
y_pred=regressor.predict(X_test)
```

```python
#Importing the r2_score function from sklearn library for evaluation.
from sklearn.metrics import r2_score 
#Calculating the r2_score with predicted data and actual test data.
r_score=r2_score(y_pred,y_test)
```

```python
#Checking the r_score value and if it is near to 1 it is a good model.
r_score
```

```python
#Importing pickle library which is used to save model in python.
import pickle
#creating a file with pickle extension .sav
filename="finalized_model_linear.sav"
#Wrting the model into the file. 'wb' represents write byte.
pickle.dump(regressor,open(filename,'wb'))
```
